```python
import matplotlib.pyplot as plt
import numpy as np

# Generating random data
np.random.seed(0)
x1 = np.arange(1, 30)
y1 = (np.random.rand(29) * 2)/10

x2 = np.arange(30, 50)
y2 = (1.5 + 0.5 * x2 + 4*np.random.rand(20))/10

x3 = np.arange(50, 70)
y3 = (20 - 0.3 * x3 + 5*np.random.rand(20))/10

x4 = np.arange(70, 90)
y4 =( 50 -0.3 * x4 - 6*np.random.rand(20))/10

# Creating subplots with a larger figure size
fig, axs = plt.subplots(figsize=(14, 6))  # Adjust the width and height as needed

# Scatter plots for each segment
axs.scatter(x1, y1, label="Freq. 1 - DS1")
axs.scatter(x2, y2, label="Freq. 1 - DS2")
axs.scatter(x3, y3, label="Freq. 1 - DS3")
axs.scatter(x4, y4, label="Freq. 1 - DS4")

# Lines of best fit
axs.plot(np.unique(x1), np.poly1d(np.polyfit(x1, y1, 0))(np.unique(x1)), color='blue')
axs.plot(np.unique(x2), np.poly1d(np.polyfit(x2, y2, 1))(np.unique(x2)), color='orange')
axs.plot(np.unique(x3), np.poly1d(np.polyfit(x3, y3, 1))(np.unique(x3)), color='green')
axs.plot(np.unique(x4), np.poly1d(np.polyfit(x4, y4, 1))(np.unique(x4)), color='red')

# Adding vertical dashed lines
axs.axvline(x=x2[0], color='gray', linestyle='--')
axs.axvline(x=x3[0], color='gray', linestyle='--')
axs.axvline(x=x4[0], color='gray', linestyle='--')

# Adding legend
axs.legend()

# Adding labels and title
axs.set_xlabel('time')
axs.set_ylabel('Hertz')

# Display the plot
plt.show()
```
