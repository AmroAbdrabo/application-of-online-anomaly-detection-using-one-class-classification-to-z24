```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
r1v = pd.read_excel("R1V.xlsx")
r2v = pd.read_excel("R2V.xlsx")
r2l = pd.read_excel("R2L.xlsx")
r2t = pd.read_excel("R2T.xlsx")
r1v.head()
```

```python
def get_fft(channel):
    frequency = channel["Frequency (Hz)"]
    amplitude = channel["Amplitude (m/s²)"]
    return {'frequency': frequency, 'amplitude': amplitude}
```

```python
def plot_fft(frequency, amplitude):
    plt.figure(figsize=(10, 6))
    plt.plot(frequency, amplitude, linestyle='-', color='blue')
    plt.title('Acceleration frequency response')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude (m/s²)')
    plt.grid(False)
    plt.show()
```

```python
plot_fft(**get_fft(r1v))
```

```python
plot_fft(**get_fft(r2v))
```

```python
plot_fft(**get_fft(r2l))
```

```python
plot_fft(**get_fft(r2t))
```
